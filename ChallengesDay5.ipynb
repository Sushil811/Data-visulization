```python
 # 3.plot a Pie Chart of Indian Car companies market share % values:.max ( 6 Companies market share )
```

```python
#importing the libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cars = ['Maruti', 'Hyundai', 'Tata',
        'Kia', 'Mahindra', 'Toyota']
  
shares_in_percent= [45.61, 16.41, 9.25, 5.96, 5.21,4.68]
 
  
# Creating plot
fig= plt.figure(figsize =(10, 7))
plt.pie(shares_in_percent, labels = cars, autopct = '%1.2f%%', shadow=True)
 
 
 
# show plot
plt.legend(title ="Cars")
plt.title('Companies market shares in %(march-21)',color='b') 
plt.axis('equal')
plt.grid(True) 
plt.show()
```

```python
# 4. Plot a Bar Chart with India's covid-19 Active cases ( Max 10 States )
```

```python
states = ['Maharashtra', 'Keral', 'Karnataka', 'Tamilnadu', 'Andhrapradesh', 'Uttarpradesh', 'Madhyapradesh', 'Haryana', 'Bihar', 'Gujarat']
```

```python
active_cases = [117698, 110616, 38752, 33665, 31850, 1789, 416, 1034, 1141, 1497]
```

```python
 
plt.barh(states,active_cases,color='yellowgreen')
plt.title ("India's covid-19 Active cases", color='g') 
plt.ylabel('States',color='b') 
plt.xlabel('Active cases',color='b')
plt.show()
```
